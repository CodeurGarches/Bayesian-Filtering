```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
def dhk(X,k,xk_0,yk_0):
    return np.array([(-(X[k][1]-yk_0)/(X[k][0]-xk_0)**2)*(1/(1+((X[k][1]-yk_0)/(X[k][0]-xk_0))**2)),(1/(X[k][0]-xk_0))*(1/(1+((X[k][1]-yk_0)/(X[k][0]-xk_0))**2)),0,0])
```

```python
T = 1 
Nb_mes = 100
phi = 135*np.pi/180
phi_opt = 0.2
V_obs = 10
X_obs_0 = [0,0,V_obs,0]
V_em = 5

P0 = np.diag([1000,1000,10,10])**2
J0 = np.linalg.inv(P0)

FI = np.array([[1,0,T,0],[0,1,0,T],[0,0,1,0],[0,0,0,1]])
Cap_em = -np.pi/9
sigma_mes = np.pi/180

X_0 = [2000,2000,V_em*np.cos(Cap_em),V_em*np.sin(Cap_em)]
```

```python
def simu(N,sigma_mes,FI,J0,X_0,X_obs_0,phi):
    J = N*[None]
    J[0] = J0
    X = N*[None]
    X[0] = np.array(X_0)
    X_obs = N*[None]
    X_obs[0] = np.array(X_obs_0)
    
    for k in range(1,N):
        #X[k] = np.matmul(FI,X[k-1])
        X[k] = FI@X[k-1]
        X_obs[k] = FI@np.array(X_obs[k-1])
        grad_hk = dhk(X,k,X_obs[k][0],X_obs[k][1])
        J[k] = (1/(sigma_mes)**2)*np.outer(grad_hk,np.transpose(grad_hk)) + (np.linalg.inv(FI).T)@J[k-1]@np.linalg.inv(FI)
        if k==51:
            X_obs[k] = np.array([X_obs[k][0], X_obs[k][1], X_obs[k][2]*np.cos(phi), X_obs[k][2]*np.sin(phi)])
            
    return X,X_obs,J
```

```python
X,X_obs,J = simu(N,sigma_mes,FI,J0,X_0,X_obs_0,phi)

X_opt,X_obs_opt,J_opt = simu(N,sigma_mes,FI,J0,X_0,X_obs_0,phi_opt)
```

```python
#Â VANILLA PHI 
J_invs=[np.linalg.inv(j) for j in J]

var_x = np.sqrt([j[0,0] for j in J_invs])
var_y = np.sqrt([j[1,1] for j in J_invs])
var_xp = np.sqrt([j[2,2] for j in J_invs])
var_yp = np.sqrt([j[3,3] for j in J_invs])

#PHI OPT 
J_invs_opt=[np.linalg.inv(j) for j in J_opt]

var_x_o = np.sqrt([j[0,0] for j in J_invs_opt])
var_y_o = np.sqrt([j[1,1] for j in J_invs_opt])
var_xp_o = np.sqrt([j[2,2] for j in J_invs_opt])
var_yp_o = np.sqrt([j[3,3] for j in J_invs_opt])
```

```python
plt.plot([x[0] for x in X_obs],[x[1] for x in X_obs])
plt.title('Trajectory 2D')
plt.xlabel('x')
plt.ylabel('y')
```

```python
plt.figure(figsize=(9,6))
plt.title('Plotting Errors in speed for vanilla and suboptimal $\Phi$')

plt.plot(range(N),var_yp,color='orchid',label='y')
plt.plot(range(N),var_yp_o,color='red',label='y_subopt')

plt.ylabel('Variance')
plt.xlabel('Measure')
plt.axvline(x=51,linestyle=':',color='navy',label='trajectory change')
plt.legend()
plt.show()
```

```python
phi_L = [k*np.pi/N for k in range(N)]
```

```python
var_y_f = []
var_x_f = []
target = []
for phi in phi_L:
    X,X_obs,J = simu(N,sigma_mes,FI,J0,X_0,X_obs_0,phi)
    J_invs=np.linalg.inv(J[-1])
    var_x_f.append(J_invs[0,0])
    var_y_f.append(J_invs[1,1])
    target.append(np.sqrt(J_invs[1,1]+J_invs[0,0]))

```

```python
plt.title('Erreur en position en fonction de $\phi$')
plt.plot(phi_L,target,color='navy')
plt.axvline(x=phi_L[target.index(min(target))],linestyle=':',color='teal')
plt.axhline(y=min(target),linestyle='--',color='blue')
plt.xlabel(r'$\phi$ (rad)')
plt.ylabel('Error (m)')
plt.show()
```
