```python
import numpy as np
import matplotlib.pyplot as plt
```

Définition des variables exemples

```python
np.random.seed(1)
```

```python
m0_ex = [5000,-20,5000,20]
sigma0_ex = [2000,20,2000,20]
c_ex = 2
n_ex = 200
dt_ex = 1
```

```python
H_ex = np.array([[1,0,0,0],[0,0,1,0]])
sigma_ex = [10,30]
```

Création des Etats grâce au système connu

```python
def F(m0,dt):
  F = np.identity(np.size(m0))
  F[0,1] = dt
  F[2,3] = dt
  return F

def G(m0,dt):
  G = np.zeros((np.size(m0),2))
  G[0,0] = dt**2/2
  G[2,1] = dt**2/2
  G[1,0] = dt
  G[3,1] = dt
  return G
```

```python
def simuEtat(dt,n,m0,sigma0,c):
  m0 = np.array(m0)
  P0 = np.zeros((len(sigma0),len(sigma0)))
  for k in range(len(sigma0)):
    P0[k,k] = sigma0[k]**2
  X0 = np.random.multivariate_normal(m0,P0)

  Q = np.identity(2)*c**2
  eps = [np.random.multivariate_normal(np.zeros(2),Q) for k in range(n)]

  F_mat = F(m0,dt)
  G_mat = G(m0,dt)

  Etat = [X0]
  for k in range(n):
    X = np.dot(F_mat,Etat[-1])+np.dot(G_mat,eps[k])
    Etat.append(X)

  return Etat
```

```python
X_ex = simuEtat(dt_ex,n_ex,m0_ex,sigma0_ex,c_ex)
xX = [X_ex[k][0] for k in range(len(X_ex))]
yX = [X_ex[k][2] for k in range(len(X_ex))]
plt.grid()
plt.title('Etats théoriques simulés du mobile')
plt.plot(xX,yX)
```

```python
T = np.linspace(0,n_ex,n_ex+1)
plt.grid()
plt.plot(T,xX,label='x positions')
plt.plot(T,yX,label='y positions')
plt.legend()
plt.title('Evolution temporelle des états simulés')
plt.show()
```

Création des Observations à partir des Etats simulés

```python
def simuObs(X,H,sigma):
  R = np.zeros((len(sigma),len(sigma)))
  for k in range(len(sigma)):
    R[k,k] = sigma[k]**2

  V = [np.random.multivariate_normal(np.zeros(len(sigma)),R) for k in range(len(X))]

  Obs = []
  for k in range(len(X)):
    Y = np.dot(H,X[k])+V[k]
    Obs.append(Y)

  return Obs
```

```python
Y_ex = simuObs(X_ex,H_ex,sigma_ex)
xY = [Y_ex[k][0] for k in range(len(X_ex))]
yY = [Y_ex[k][1] for k in range(len(X_ex))]
plt.grid()
plt.title('Observations bruités des positions du mobile')
plt.plot(xY,yY)
```

```python
plt.grid()
plt.plot(xX,yX,label='Etats simulés')
plt.plot(xY,yY,label='Observations')
plt.legend()
plt.title('Evolution spatiale du mobile')
plt.show()
```

Création du filtre de Kalman grâce aux observations uniquement

```python
def Kalman(Y,dt,m0,sigma0,c,H,sigma):
  P0 = np.zeros((len(sigma0),len(sigma0)))
  for k in range(len(sigma0)):
    P0[k,k] = sigma0[k]**2
  R = np.zeros((len(sigma),len(sigma)))
  for k in range(len(sigma)):
    R[k,k] = sigma[k]**2
  Q = np.identity(2)*c**2

  F_mat = F(m0,dt)
  G_mat = G(m0,dt)

  Xchap = [m0]
  Pchap = [P0]
  for k in range(len(Y)):
    Xpred = np.dot(F_mat,Xchap[-1])
    Ppred = np.dot(np.dot(F_mat,Pchap[-1]),np.transpose(F_mat)) + np.dot(np.dot(G_mat,Q),np.transpose(G_mat))

    I = np.dot(np.dot(H,Ppred),np.transpose(H)) + R
    K = np.dot(np.dot(Ppred,np.transpose(H)),np.linalg.inv(I))

    Xcor = Xpred + np.dot(K,Y[k] - np.dot(H,Xpred))
    Pcor = np.dot(np.identity(len(m0)) - np.dot(K,H),Ppred)

    Xchap.append(Xcor)
    Pchap.append(Pcor)

  return Xchap, Pchap
```

```python
Z_ex = Kalman(Y_ex,dt_ex,m0_ex,sigma0_ex,c_ex,H_ex,sigma_ex)[0]
xZ = [Z_ex[k][0] for k in range(len(Z_ex))][1::]
yZ = [Z_ex[k][2] for k in range(len(Z_ex))][1::]
plt.grid()
plt.title('Estimation des positions du mobile')
plt.plot(xZ,yZ)
```

Définition du système non linéaire

```python
def h(X):
    return np.array([np.sqrt(X[0]**2+X[2]**2),np.arctan(X[2]/X[0])])

def gradh(X):
    Z = np.zeros((2,2))
    Z[0,0] = X[0]/np.sqrt(X[0]**2+X[2]**2)
    Z[0,1] = X[2]/np.sqrt(X[0]**2+X[2]**2)
    Z[1,0] = -X[2]/(X[0]**2+X[2]**2)
    Z[1,1] = X[0]/(X[0]**2+X[2]**2)
    return Z
```

Création des Observations dans le cas étendu

```python
def simuEtatEtendu(X):
  Etats = [h(X[k]) for k in range(len(X))]
  return Etats
```

```python
def simuObsEtendu(X,sigma):
  R = np.zeros((len(sigma),len(sigma)))
  for k in range(len(sigma)):
    R[k,k] = sigma[k]**2
  V = [np.random.multivariate_normal(np.zeros(len(sigma)),R) for k in range(len(X))]

  Obs = []
  for k in range(len(X)):
      Y = h(X[k])+V[k]
      Obs.append(Y)

  return Obs
```
